```python
data = [[150, 50],
        [150, 60],
        [150, 70],
        [155, 70],
        [160, 70],
        [170, 70],
        [175, 70],
        [180, 70],
        [175, 80],
        [180, 90]
        ]

homework_data = [[160, 50],
                [165, 60],
                [170, 80],
                 [175, 90],
                ]
```

```python
import numpy as np
```

```python
data_np = np.array(data)
homework_data_np = np.array(homework_data)
```

```python
data_np.shape
```

```python
data_np[2, 1]
```

```python
x = data_np[:, 0]
y = data_np[:, 1]
x_h = homework_data_np[:, 0]
y_h = homework_data_np[:, 1]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x, y)
plt.xlabel("Height[cm]")
plt.ylabel("Weight[kg]")
plt.grid(linestyle="--")
n = data_np.shape[0]
s_xx = np.sum(x * x) / n
s_xy = np.sum(x * y) / n
s_x = np.sum(x) / n
s_y = np.sum(y) / n
a = (s_xy - s_x * s_y) / (s_xx - s_x * s_x)
b = s_y - a * s_x
y_hat = a * x + b
plt.plot(x, y_hat)
```

```python
plt.scatter(x, y)
plt.scatter(x_h, y_h, color="red")
plt.xlabel("Height[cm]")
plt.ylabel("Weight[kg]")
plt.grid(linestyle="--")
n = data_np.shape[0]
s_xx = np.sum(x * x) / n
s_xy = np.sum(x * y) / n
s_x = np.sum(x) / n
s_y = np.sum(y) / n
a = (s_xy - s_x * s_y) / (s_xx - s_x * s_x)
b = s_y - a * s_x
y_hat = a * x + b
plt.plot(x, y_hat)
```
